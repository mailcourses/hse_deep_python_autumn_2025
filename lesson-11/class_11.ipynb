```python
def create_logger(name):
    logger = logging.getLogger(name)
    logger.setLevel(logging.INFO)

    format1 = logging.Formatter("FILE %(asctime)s\t%(levelname)s\t%(name)s\t%(message)s")
    format2 = logging.Formatter("STREAM %(asctime)s\t%(levelname)s\t%(name)s\t%(message)s")

    file_handler = logging.FileHandler(f"{name}.log", mode="w")
    file_handler.setLevel(logging.WARNING)
    file_handler.setFormatter(format1)

    stream_handler = logging.StreamHandler()
    stream_handler.setLevel(logging.DEBUG)
    stream_handler.setFormatter(format2)

    logger.addHandler(file_handler)
    logger.addHandler(stream_handler)

    return logger
```

```python
def create_logger(name):
    logger = logging.getLogger(name)
    logger.setLevel(logging.INFO)

    #if logger.handlers:
    #    return logger

    while logger.handlers:
        logger.removeHandler(logger.handlers[-1])
    
    format1 = logging.Formatter("FILE %(asctime)s\t%(levelname)s\t%(name)s\t%(message)s")
    format2 = logging.Formatter("STREAM %(asctime)s\t%(levelname)s\t%(name)s\t%(message)s")

    file_handler = logging.FileHandler(f"{name}.log", mode="w")
    file_handler.setLevel(logging.WARNING)
    file_handler.setFormatter(format1)

    stream_handler = logging.StreamHandler()
    stream_handler.setLevel(logging.DEBUG)
    stream_handler.setFormatter(format2)

    logger.addHandler(file_handler)
    logger.addHandler(stream_handler)

    return logger
```

```python
class CustomFilter(logging.Filter):
    def filter(self, rec):
        return rec.args[0] != "steve"


while book.filters:
    book.removeFilter(book.filters[-1])

book.addFilter(CustomFilter())
```
